```python
import pandas as pd
import glob
import os

# Carpeta donde est√°n los CSV
carpeta = "C:/Users/Nico/Desktop/Seguimiento"

# Buscar todos los CSV en la carpeta
archivos = glob.glob(os.path.join(carpeta, "*.csv"))

dfs = []

for archivo in archivos:
    df_temp = pd.read_csv(archivo)
    df_temp["archivo"] = os.path.basename(archivo)  # Agregar nombre del archivo
    dfs.append(df_temp)

# Concatenar
df = pd.concat(dfs, ignore_index=True)

# Mostrar una vista general
print(f"Se cargaron {len(archivos)} archivos con {len(df)} filas en total.")
print(df.head())

df.to_csv(os.path.join(carpeta, "resultados_modelos.csv"), index=False)
```
