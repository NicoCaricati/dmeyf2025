```python
import duckdb

path_02 = "../../../buckets/b1/datasets/competencia_02_crudo.csv.gz"
path_03 = "../../../buckets/b1/datasets/competencia_03_crudo.csv.gz"
path_04 = "../../../buckets/b1/datasets/competencia_04_crudo.csv.gz"

con = duckdb.connect()

# Concatenar directamente: 02 con header, 03 sin header (saltando la primera fila)
con.execute(f"""
    COPY (
        SELECT * FROM read_csv_auto('{path_02}', header=True, compression='gzip')
        UNION ALL
        SELECT * FROM read_csv_auto('{path_03}', header=False, skip=1, compression='gzip')
    ) TO '{path_04}' (FORMAT CSV, COMPRESSION GZIP, HEADER TRUE)
""")

con.close()
```

```python
path_02 = "../../../buckets/b1/datasets/competencia_02_crudo.csv.gz"
path_03 = "../../../buckets/b1/datasets/competencia_03_crudo.csv.gz"
path_04 = "../../../buckets/b1/datasets/competencia_04_crudo.csv.gz"

con = duckdb.connect()

# Leer 02 con header
df02 = con.execute(f"""
    SELECT * FROM read_csv_auto('{path_02}', header=True, compression='gzip')
""").fetchdf()

# Leer 03 sin header (saltando primera fila)
df03 = con.execute(f"""
    SELECT * FROM read_csv_auto('{path_03}', header=True, compression='gzip')
""").fetchdf()

# Leer 04 (resultado final)
df04 = con.execute(f"""
    SELECT * FROM read_csv_auto('{path_04}', header=True, compression='gzip')
""").fetchdf()

# --- CHECKS ---
print("üîé Columnas en 02:", list(df02.columns))
print("üîé Columnas en 03:", list(df03.columns))
print("üîé Columnas en 04:", list(df04.columns))

# Check columnas iguales entre 02 y 03
if list(df02.columns) == list(df03.columns):
    print("‚úÖ Las columnas de 02 y 03 coinciden")
else:
    print("‚ùå Las columnas difieren entre 02 y 03")

# Check cantidad de filas
n02 = len(df02)
n03 = len(df03)
n04 = len(df04)

print(f"üìä Filas en 02: {n02}")
print(f"üìä Filas en 03 (sin header): {n03}")
print(f"üìä Filas en 04: {n04}")

if n04 == n02 + n03:
    print("‚úÖ La cantidad de filas en 04 es correcta")
else:
    print("‚ùå La cantidad de filas en 04 NO coincide con la suma esperada")
```
