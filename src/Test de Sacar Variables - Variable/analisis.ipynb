```python
import pandas as pd
import glob
import os

# Carpeta donde est√°n los CSV
carpeta = "C:/Users/Nico/Desktop/Test de Sacar Variables - Copy"

# Buscar todos los CSV en la carpeta
archivos = glob.glob(os.path.join(carpeta, "*.csv"))

# Leer y concatenar todos los CSV
df = pd.concat([pd.read_csv(archivo) for archivo in archivos], ignore_index=True)

# Mostrar una vista general
print(f"Se cargaron {len(archivos)} archivos con {len(df)} filas en total.")
print(df.head())

df.to_csv(os.path.join(carpeta, "datos_combinados.csv"), index=False)
```
